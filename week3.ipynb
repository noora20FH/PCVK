#Linier Transformation

```python
print('Vhanges The Brightness Level')
print('----------------------------')
try:
  brightness = int(input('Input The value of Brightness: '))
except ValueError:
  print('Error, Not a nuber')


brightness_image = np.zeros(original1.shape, original1.dtype)

#access each pixel
for y in range(original1.shape[0]):
  for x in range(original1.shape[1]):
    for c in range(original1.shape[2]):
                                #NumPy function that is used to limit the values of an array within a specified range
                                #np.clip(a, a_min, a_max, out=None)
      brightness_image[y,x,c] = np.clip(original1[y, x, c] + brightness, 0, 255) #np.clip(......, 0, 255)already same with the truncate, below 0 becom o; higher than 255 become 255

#Simple way without looping
#brightness_image = cv.convertScaleAbs(original, beta=brightness)

final_frame = cv.hconcat((original1, brightness_image))
cv2_imshow(final_frame)
```

```python

img = cv.imread('/content/drive/MyDrive/mandrill.tiff')
image = cv.cvtColor(img,cv.COLOR_BGR2RGB) # konversi channel BGR -> RGB
plt.imshow(image)
```

```python
print('Implement Image Invert')
print('----------------------------')


inverted_image = np.zeros(image.shape, img.dtype)

#access each pixel
for y in range(image.shape[0]):
  for x in range(image.shape[1]):
    for c in range(image.shape[2]):
      inverted_image[y,x,c] = np.clip(255 - image[y, x, c] , 0, 255)

final_frame = cv.hconcat((image, inverted_image))
plt.imshow(final_frame)
```

#Implement transformation of logarithmic brightness on Google Colaboratory using the formula written in the Theoretical Background Reviews, generating output shown in

```python
print('Mengubah tingkat kecerahan citra dengan Transformasi Log')
print('--------------------------------------------------------')
try:
  brightness = int(input('Masukkan nilai kecerahan: '))
except ValueError:
  print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/female.tiff')
brightness_image = np.zeros(original.shape, original.dtype)

# akses per piksel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(brightness * (np.log(1 + original[y,x,c])), 0, 255)

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

# Implement grayscale transformation using the averaging, lightness, and luminance method on Google Colaboratory using the formula written in the Theoretical Background Reviews, generating output as follows 

```python

img_m = cv.imread('/content/drive/MyDrive/mandrill.tiff')
img_rgb = cv.cvtColor(img_m,cv.COLOR_BGR2RGB) # konversi channel BGR -> RGB
plt.imshow(img_rgb)
     
```

## Averaging

mean function in numpy

```python
print('Implement Image Invert')
print('----------------------------')



#access each pixel
for y in range(img_rgb.shape[0]):
  for x in range(img_rgb.shape[1]):
    for c in range(img_rgb.shape[2]):

      averaged_img[y,x,c] = np.mean(img_rgb[y,x, : ])


final_frame = cv.hconcat((img_rgb, averaged_img))
plt.imshow(final_frame)
```

##Lightness

```python


lightness_img = np.zeros(img_rgb.shape, img_rgb.dtype)

#access each pixel
for y in range(img_rgb.shape[0]):
  for x in range(img_rgb.shape[1]):
    for c in range(img_rgb.shape[2]):

     lightness_img[y,x,c] = np.max(img_rgb[y,x, : ])/2 + np.min(img_rgb[y,x, : ])/2


final_frame = cv.hconcat((img_rgb, lightness_img))
plt.imshow(final_frame)
```

##Luminance

```python
luminance_img = np.zeros(img_rgb.shape, img_rgb.dtype)
# luminosity_constant = [0.21,0.72,0.07]
#access each pixel
for y in range(img_rgb.shape[0]):
  for x in range(img_rgb.shape[1]):
    for c in range(img_rgb.shape[2]):

     luminance_img[y,x,c] = 0.21*img_rgb[y,x, 0 ] + 0.72*img_rgb[y,x,1 ] + 0.07*img_rgb[y,x, 2 ]


final_frame = cv.hconcat((img_rgb, luminance_img))
plt.imshow(final_frame)
```
