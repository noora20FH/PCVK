#Creating 2D image

```python
black_image = np.zeros(shape=(513,512,3),dtype=np.int16) #creating dimensonal matrix all 0 for each pixcel
#shape(height,width,layerRGB)
plt.imshow(black_image)
```

```python
cv.rectangle(black_image,pt1=(384,0), pt2=(510,150), color=(0,255,0), thickness=10) #(x,y)
#pt1 = coordinate of UPPER LEFT corner
#pt2 = coordinate of BUTTOM LEFT corner
plt.imshow(black_image)
```

```python
cv.rectangle(black_image, pt1=(200,200), pt2=(300,300), color=(0,0,255), thickness = 15)
plt.imshow(black_image)
```

```python
cv.circle(black_image, center=(100,100), radius=50, color=(255,0,0), thickness = 8)
plt.imshow(black_image)
```

```python
cv.line(black_image, pt1=(0,0),pt2=(512,512), color=(255,255,255), thickness=5)
plt.imshow(black_image)
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_image, text='Noora', org=(10,500), fontFace=font, fontScale=4, color=(255,255,0), thickness=2,lineType=cv.LINE_AA )
#lineType=line connectivity
plt.imshow(black_image)
```

*   What is the difference between the images displayed without and with matplotlib?
  - without using matplotlib the image that appears will be in accordance with the original size. Meanwhile, if you use matplotlib, the image will be adjusted to the plot that has been scaled

*   What is the difference and effect of creating a black image between int16 and int32 data types?
  - int16 memiliki rentang nilai -32768 to 32767,sedangkan int32 memiliki rentang nilai yang lebih besar yaitu  -2147483648 to 2147483647. Berdasarkan nilai tersebut, semakin besar rentan value maka gambar yang dihasilkan dapat lebih presisi.
