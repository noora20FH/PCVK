- use NumPy indexing to copy the green and blue channel data to the appropriate positions in the new green_blue_image array.
-  this method only displays the green and blue channels of the image, and sets the red channel to zero.

```python
#use NumPy indexing
#index 0 = red; 1=green, 2= blue

# Extract green and blue channels
green_blue_image = np.zeros_like(img6)
green_blue_image[:, :, 1] = img6[:, :, 1]  # Green channel
green_blue_image[:, :, 2] = img6[:, :, 2]  # Blue channel

# Display green-blue image using plt.imshow
plt.imshow(green_blue_image)
plt.show()
```

##Create 5 boxes of different sizes and colors in one image. it is recommended to use random numbers

```python
amphibi = cv.imread('/content/drive/MyDrive/amphibians1.png')
amphibi1 = amphibi
amphibi1 = cv.cvtColor(amphibi,cv.COLOR_BGR2RGB)
plt.imshow(amphibi1)
tinggi = amphibi1.shape[0]
lebar = amphibi1.shape[1]
print("reslusi image: tinggixlebar = ", tinggi,' x ', lebar)
```

```python
import random

amphibi2 = amphibi1

for i in range(5):#generate 5 random boxes
  x = np.random.randint(0, 512 - 50)  # Random x coordinate, the pixel
  y = np.random.randint(0, 512 - 50)  # Random y coordinate
  w = np.random.randint(10, 35)       # Random width, the size of the box
  h = np.random.randint(10, 35)       # Random height
  r = np.random.randint(0, 256)       # Random red value
  g = np.random.randint(0, 256)       # Random green value
  b = np.random.randint(0, 256)       # Random blue value
  color = (r, g, b)                   # Random color
    
    # Draw the box on the image
  amphibi2[y:y+h, x:x+w, :] = color

# Display the image
plt.imshow(amphibi2)
plt.show()
```
